## Theory Questions

### Theory 1

**Q1. What is Object-Oriented Programming (OOP)?**

**Answer:** **OOP** is a programming paradigm that organizes code around **objects** (data + behavior). It models real-world entities by combining **attributes** (data) and **methods** (functions) into reusable units, improving modularity, reusability, and maintainability.

### Theory 2

**Q2. What is a class in OOP?**

**Answer:** A **class** is a blueprint/template that defines the structure (attributes) and behavior (methods) that its objects will have.

### Theory 3

**Q3. What is an object in OOP?**

**Answer:** An **object** is an instance of a class. It has its own identity and its own values for the attributes defined by the class.

### Theory 3

**Q3. What is an object in OOP?**

### Theory 4

**Q4. What is the difference between abstraction and encapsulation?**

**Answer:** **Abstraction** focuses on *what* an object does (exposing essential features and hiding unnecessary details). **Encapsulation** bundles data and methods together and controls access to internal state (e.g., via private/protected members and properties).

### Theory 5

**Q5. What are dunder methods in Python?**

**Answer:** **Dunder (double underscore) methods** like `__init__`, `__str__`, `__len__` are special methods that let you define/override built‑in behavior for your classes (operator overloading, printing, iteration, etc.).

### Theory 6

**Q6. Explain the concept of inheritance in OOP.**

**Answer:** **Inheritance** allows a class (child/derived) to acquire attributes and methods from another class (parent/base), enabling code reuse and hierarchical relationships.

### Theory 7

**Q7. What is polymorphism in OOP?**

**Answer:** **Polymorphism** means “many forms”: the same interface/method name can behave differently depending on the object’s type (e.g., overridden methods in subclasses).

### Theory 8

**Q8. How is encapsulation achieved in Python?**

**Answer:** In Python, encapsulation is typically achieved by:
- Naming conventions: `_protected` and `__private` (name mangling)
- Properties (`@property`) to control attribute access
- Providing methods to read/update internal state safely.

### Theory 9

**Q9. What is a constructor in Python?**

**Answer:** A **constructor** initializes a new object. In Python, the constructor is `__init__(self, ...)`, called after the object is created.

### Theory 10

**Q10. What are class and static methods in Python?**

**Answer:** **Class methods** use `@classmethod` and receive `cls` (the class) to act on class-level state. **Static methods** use `@staticmethod` and receive no automatic `self/cls`; they behave like regular functions placed inside a class namespace.

### Theory 11

**Q11. What is method overloading in Python?**

**Answer:** Python doesn’t support true compile-time method overloading by signature. If you define the same method name multiple times, the latest definition wins. Overloading is usually simulated using default arguments, `*args/**kwargs`, or `functools.singledispatch`.

### Theory 12

**Q12. What is method overriding in OOP?**

**Answer:** **Method overriding** is when a subclass provides its own implementation of a method already defined in the parent class (same method name), changing/extending behavior.

### Theory 13

**Q13. What is a property decorator in Python?**

**Answer:** `@property` lets you access a method like an attribute. It’s used to create **managed attributes** with getters/setters/deleters while keeping a clean attribute-style interface.

### Theory 14

**Q14. Why is polymorphism important in OOP?**

**Answer:** It allows writing code that works with objects through a common interface (e.g., calling `fly()` on any Bird), improving flexibility and enabling extensibility without modifying existing code.

### Theory 15

**Q15. What is an abstract class in Python?**

**Answer:** An **abstract class** is a class meant to be subclassed, often containing **abstract methods** that must be implemented by concrete subclasses. In Python this is typically done with `abc.ABC` and `@abstractmethod`.

### Theory 16

**Q16. What are the advantages of OOP?**

**Answer:** Key advantages: modular design, reusability (inheritance), maintainability, easier debugging/testing, data hiding (encapsulation), and the ability to model real-world entities naturally.

### Theory 17

**Q17. What is the difference between a class variable and an instance variable?**

**Answer:** A **class variable** is shared by all instances of the class (stored on the class). An **instance variable** belongs to a particular object (stored on that object).

### Theory 18

**Q18. What is multiple inheritance in Python?**

**Answer:** **Multiple inheritance** is when a class inherits from more than one base class. Python supports it and resolves method/attribute lookup using the MRO (Method Resolution Order).

### Theory 19

**Q19. Explain the purpose of '__str__' and '__repr__' methods in Python.**

**Answer:** `__str__` returns a human-friendly string (used by `print()` and `str()`). `__repr__` returns an unambiguous/debug-focused representation (used by `repr()` and the interactive prompt). Ideally, `__repr__` can recreate the object when possible.

### Theory 20

**Q20. What is the significance of the 'super()' function in Python?**

**Answer:** `super()` gives access to methods of the parent class in a cooperative way, respecting Python’s MRO. It’s commonly used to call parent implementations inside overrides (especially in multiple inheritance).

### Theory 21

**Q21. What is the significance of the '__del__' method in Python?**

**Answer:** `__del__` is a finalizer called when an object is about to be garbage-collected. It’s not guaranteed to run immediately (or at all at interpreter shutdown), so it shouldn’t be used for critical resource management—use context managers instead.

### Theory 22

**Q22. What is the difference between @staticmethod and @classmethod in Python?**

**Answer:** `@staticmethod` defines a method that doesn’t receive `self` or `cls`. `@classmethod` receives the class (`cls`) and can modify class state or act as alternative constructors.

### Theory 23

**Q23. How does polymorphism work in Python with inheritance?**

**Answer:** In inheritance, polymorphism works when a base-class reference points to a subclass object; calling an overridden method executes the subclass’s version at runtime (dynamic dispatch).

### Theory 24

**Q24. What is method chaining in Python OOP?**

**Answer:** Method chaining returns `self` (or another object) from a method so that calls can be chained: `obj.set_x(1).set_y(2).run()`.

### Theory 25

**Q25. What is the purpose of the '__call__' method in Python?**

**Answer:** `__call__` makes an instance callable like a function. If a class defines `__call__`, you can do `obj()` and it runs `obj.__call__()`.

### Practical 2

**Q2. 2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.**

**Solution:**

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height

print("Circle area:", Circle(3).area())
print("Rectangle area:", Rectangle(4, 5).area())
```

**Q3. 3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.**

**Solution:**

```python
class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type

class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery_kwh):
        super().__init__(vehicle_type, brand)
        self.battery_kwh = battery_kwh

ecar = ElectricCar("Car", "Tesla", 75)
print(ecar.type, ecar.brand, ecar.battery_kwh, "kWh")
```

### Practical 5

**Q5. 5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.**

**Solution:**

```python
class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = float(initial_balance)  # private (name-mangled)

    def deposit(self, amount):
        amount = float(amount)
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")
        self.__balance += amount

    def withdraw(self, amount):
        amount = float(amount)
        if amount <= 0:
            raise ValueError("Withdraw amount must be positive.")
        if amount > self.__balance:
            raise ValueError("Insufficient balance.")
        self.__balance -= amount

    def check_balance(self):
        return self.__balance

acct = BankAccount(1000)
acct.deposit(500)
acct.withdraw(300)
print("Balance:", acct.check_balance())
```

### Practical 6

**Q6. 6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().**

**Solution:**

```python
class Instrument:
    def play(self):
        raise NotImplementedError("Subclasses must implement play().")

class Guitar(Instrument):
    def play(self):
        print("Guitar: strum strum!")

class Piano(Instrument):
    def play(self):
        print("Piano: plink plonk!")

def perform(instrument: Instrument):
    instrument.play()

perform(Guitar())
perform(Piano())
```

**Solution:**

```python
class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be 0.")
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

f = Fraction(3, 4)
print(str(f))
```

### Practical 10

**Q10. 10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.**

**Solution:**

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if not isinstance(other, Vector):
            return NotImplemented
        return Vector(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Vector(x={self.x}, y={self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(v1 + v2)
```

**Q12. 12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.**

**Solution:**

```python
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = list(grades)

    def average_grade(self):
        if not self.grades:
            return 0.0
        return sum(self.grades) / len(self.grades)

s = Student("Alex", [85, 90, 78, 92])
print("Average:", s.average_grade())
```

### Practical 14

**Q14. 14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.**

**Solution:**

```python
class Employee:
    def __init__(self, hours_worked, hourly_rate):
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, hours_worked, hourly_rate, bonus):
        super().__init__(hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.bonus

e = Employee(40, 500)
m = Manager(40, 500, 5000)
print("Employee salary:", e.calculate_salary())
print("Manager salary:", m.calculate_salary())
```

### Practical 16

**Q16. 16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.**

**Solution:**

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        return "Moo!"

class Sheep(Animal):
    def sound(self):
        return "Baa!"

print("Cow:", Cow().sound())
print("Sheep:", Sheep().sound())
```
